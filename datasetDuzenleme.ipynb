```python
import pandas as pd
import os

# Yeni dosya yolları
csv_path = "C:/Users/pc/Desktop/DiaRet/step2/train.csv"
image_dir = "C:/Users/pc/Desktop/DiaRet/step2/train_images"

# CSV dosyasını oku
df = pd.read_csv(csv_path)

# Label'i 0 olanları filtrele
label_0_df = df[df['diagnosis'] == 0]

# Kaç adet olduğunu yazdır
print(f"Silinecek görsel sayısı: {len(label_0_df)}")

# Görselleri sil
for img_id in label_0_df['id_code']:
    img_path = os.path.join(image_dir, f"{img_id}.png")
    if os.path.exists(img_path):
        os.remove(img_path)

print("0 sınıfına ait görseller silindi.")

# CSV'den 0 sınıfını sil
df = df[df['diagnosis'] != 0]
df.to_csv(csv_path, index=False)

print("0 sınıfına ait kayıtlar CSV dosyasından da silindi.")
```

```python
csv_path = "C:/Users/pc/Desktop/DiaRet/step2/train.csv"
df = pd.read_csv(csv_path)

# diagnosis: 2 → 0, diğerleri → 1
df['diagnosis'] = df['diagnosis'].apply(lambda x: 0 if x == 2 else 1)

# Güncellenmiş dağılımı yazdır
print("Sınıf dağılımı:")
print(df['diagnosis'].value_counts())

# Değişiklikleri tekrar kaydet
df.to_csv(csv_path, index=False)
```
